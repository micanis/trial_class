```
import numpy as np
import matplotlib.pyplot as plt
```

## データの生成

```
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

## グラフの描画

```
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
